```python
import pandas as pd
```

```python
df = pd.read_csv('BACE1_01_bioactivity_data_raw.csv')
```

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3.reset_index(drop=True, inplace=True)
df4 = df3
df4
```

```python
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

```python
df_no_smiles = df5.drop(columns='canonical_smiles')
df_no_smiles
```

```python
smiles = []

for i in df5.canonical_smiles.tolist():
  cpd = str(i).split('.')
  cpd_longest = max(cpd, key = len)
  smiles.append(cpd_longest)


smiles = pd.Series(smiles, name = 'canonical_smiles')
```

```python
df_clean_smiles = pd.concat([df_no_smiles,smiles], axis=1)
df_clean_smiles
```

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df_lipinski = lipinski(df_clean_smiles.canonical_smiles)
df_lipinski
```

```python
df_combined = pd.concat([df5,df_lipinski], axis=1)
```

```python
df_combined
```

```python
def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value_norm', 1)
        
    return x
```

```python
df_combined['standard_value'] = df_combined.standard_value.astype(float)
df_combined.standard_value.describe()
```

```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('standard_value', 1)
        
    return x
```

```python
df_norm = norm_value(df_combined)
df_norm
```

```python
df_norm.standard_value_norm.describe()
```

```python
df_final = pIC50(df_norm)
df_final
```

```python
df_final.pIC50.describe()
```

```python
# add class info

def pic50_to_2class(pic50):
    if pic50 < 7:
        return 0
    else:
        return 1
```

```python
df_final['class'] = df_final['pIC50'].apply(pic50_to_2class)
```

```python
import os
import deepchem as dc
from deepchem.molnet.load_function.molnet_loader import TransformerGenerator, _MolnetLoader
from deepchem.data import Dataset
from typing import List, Optional, Tuple, Union
```

```python
# load the dataset

chembl_bace_URL = 'chembl_bace_class_and_pic50_20210802_temp.csv'
chembl_bace_TASKS = ['class']

class _chembl_bace_PLoader(_MolnetLoader):

  def create_dataset(self) -> Dataset:
    loader = dc.data.CSVLoader(
        tasks=self.tasks, feature_field="canonical_smiles", featurizer=self.featurizer)
    return loader.create_dataset(dataset_file, shard_size=8192)


def load_chembl_bace(
    featurizer: Union[dc.feat.Featurizer, str] = 'ECFP',
    splitter: Union[dc.splits.Splitter, str, None] = 'scaffold',
    transformers: List[Union[TransformerGenerator, str]] = ['balancing'],
    reload: bool = True,
    data_dir: Optional[str] = None,
    save_dir: Optional[str] = None,
    **kwargs
) -> Tuple[List[str], Tuple[Dataset, ...], List[dc.trans.Transformer]]:
  """Load chembl_bace dataset
 

  """
  loader = _chembl_bace_PLoader(featurizer, splitter, transformers, chembl_bace_TASKS, data_dir,
                       save_dir, **kwargs)
  return loader.load_dataset('chembl_bace1', reload)
```

```python
chembl_bace_tasks, chembl_bace_datasets, transformers = load_chembl_bace(featurizer='ECFP', splitter='scaffold')
train_dataset, valid_dataset, test_dataset = chembl_bace_datasets
```

```python
class ClassificationModel(tf.keras.Model):
    
    def __init__(self):
        super(ClassificationModel, self).__init__()
        self.dense1 = tf.keras.layers.Dense(1000, activation='relu')
        self.dense2 = tf.keras.layers.Dense(1)

    def call(self, inputs, training=False):
        y = self.dense1(inputs)
        if training:
            y = tf.nn.dropout(y, 0.5)
        logits = self.dense2(y)
        output = tf.nn.sigmoid(logits)
        return output, logits
```

```python
keras_model = ClassificationModel()
output_types = ['prediction', 'loss']
model = dc.models.KerasModel(keras_model, dc.models.losses.SigmoidCrossEntropy(), output_types=output_types)
```

```python
model.fit(train_dataset, nb_epoch=100)
metric = dc.metrics.Metric(dc.metrics.roc_auc_score)
print('training set score:', model.evaluate(train_dataset, [metric]))
print('validation set score:', model.evaluate(valid_dataset, [metric]))
print('test set score:', model.evaluate(test_dataset, [metric]))
```
